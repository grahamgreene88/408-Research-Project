```python
import pandas as pd
import numpy as np
```

## Preprocessing - Removing Extra Rows and Columns

```python
# Importing MSA dataframe
MSA_df = pd.read_csv('MSA_Dataframe.csv', index_col=0)
print(MSA_df)
```

```python
# Removing reference PA14 strain columns
MSA_df = MSA_df.drop('Reference PA14')
print(MSA_df)
```

```python
MSA_df = MSA_df.drop('UCBPP-PA14')
print(MSA_df)
```

```python
# Dropping delta G column
MSA_df = MSA_df.drop('Delta G', axis='columns')
print(MSA_df)
```

## Preprocessing - Adding resistance phenotype column to dataframe

```python
# Importing phenotype data
phenotype_df = pd.read_excel('IsolatePhenotypes.xlsx', index_col=0)
# Merge phenotype dataframe with original MSA dataframe
ML_df = MSA_df.merge(phenotype_df, left_index=True, right_index=True)
print(ML_df)
```

```python
# replace 'S' with 0 and 'R' with 1 to make labels binary
ML_df['Phenotype'] = ML_df['Phenotype'].replace({'S': 0, 'R': 1})
print(ML_df)
```

```python
# Sort index alphabetically
ML_df_sorted = ML_df.sort_index()
pd.set_option('display.max_rows', None)
print(ML_df_sorted)
```

```python
pd.reset_option('max_columns')
pd.reset_option('max_rows')
```

## Machine Learning

```python
# Importing classifiers from sklearn
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score, roc_curve, auc
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import make_scorer, f1_score
from sklearn.calibration import CalibratedClassifierCV
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Splicing the dataframe
# iloc selects every row and every column except last column (which is the labels)
X = ML_df.iloc[:,:-1].values
y = ML_df['Phenotype']
```

```python
# Randomly split the data into training and testing sets
# Specify the stratify parametre as y to ensure that the proportion of S to R strains in each set
# is similar to the original data set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
```

```python
# Cross validation strategy
cv = StratifiedKFold(n_splits=5)
```

```python
# Defining the classifiers
# L2 penalty for logreg model results in slightly better accruacy, F1 and auc
# L1, L2 penaly for linearSVC have same metrics
# linear kernel was best for SBC classifier (better than rbf, poly and sigmoid)
logreg_model = LogisticRegression(penalty='l2', solver='liblinear')
linearSVC_model = LinearSVC(penalty='l1', dual=False, max_iter=100000)
SVC_model = SVC(kernel='linear', probability=True)
RF_model = RandomForestClassifier()
```

```python
# Perform cross validation
logreg_scores = cross_val_score(logreg_model, X_train, y_train, cv=cv)
linearSVC_scores = cross_val_score(linearSVC_model, X_train, y_train, cv=cv)
SVC_scores = cross_val_score(SVC_model, X_train, y_train, cv=cv)
RF_scores = cross_val_score(RF_model, X_train, y_train, cv=cv)
```

```python
# Displaying accuracy score of classifiers in cross validation
# Accuracy score is the ratio of correct predictions to total predictions
print('Accuracy logistic regression scores:', logreg_scores)
print('Mean logistic regression accuracy scores:', logreg_scores.mean())

print('\nLinear SVC accuracy scores:', linearSVC_scores)
print('Mean linear SVC accuracy scores:', linearSVC_scores.mean())

print('\nSVC accuracy scores:', SVC_scores)
print('Mean SVC accuracy scores:', SVC_scores.mean())

print('\nRandom forest accuracy scores:', RF_scores)
print('Mean random forest accuracy scores:', RF_scores.mean())
```

```python
# Perform cross validation for F1-macro scores
# f1_scorer = make_scorer(f1_macro, pos_label='R')
logreg_F1_scores = cross_val_score(logreg_model, X, y, cv=cv, scoring='f1_macro')
linearSVC_F1_scores = cross_val_score(linearSVC_model, X, y, cv=cv, scoring='f1_macro')
SVC_F1_scores = cross_val_score(SVC_model, X, y, cv=cv, scoring='f1_macro')
RF_F1_scores = cross_val_score(RF_model, X, y, cv=cv, scoring='f1_macro')
```

```python
# Displaying F1-macro score of classifiers
# Precision is proportion of true positives over total positive predictions
# Recall is proportion of true positive predictions over total number of actual positives
# F1-score is the mean of precision and recall
print('Logistic regression F1-macro scores:', logreg_F1_scores)
print('Mean logistic regression F1-macro scores:', logreg_F1_scores.mean())

print('\nLinear SVC F1-macro scores:', linearSVC_F1_scores)
print('Mean linear SVC F1-macro scores:', linearSVC_F1_scores.mean())

print('\nSVC F1-macro scores:', SVC_F1_scores)
print('Mean SVC F1-macro scores:', SVC_F1_scores.mean())

print('\nRandom forest F1-macro scores:', RF_F1_scores)
print('Mean random forest F1-macro scores:', RF_F1_scores.mean())
```

```python
# Perform cross validation for roc_auc scores
logreg_auc_scores = cross_val_score(logreg_model, X, y, cv=cv, scoring='roc_auc')
linearSVC_auc_scores = cross_val_score(linearSVC_model, X, y, cv=cv, scoring='roc_auc')
SVC_auc_scores = cross_val_score(SVC_model, X, y, cv=cv, scoring='roc_auc')
RF_auc_scores = cross_val_score(RF_model, X, y, cv=cv, scoring='roc_auc')
```

```python
# Displaying roc_auc score of classifiers
# roc is tradeoff between recall and specificity - perfect is 1.0 - random classifier is 0.5
print('Logistic regression auc scores:', logreg_auc_scores)
print('Mean logistic regression auc scores:', logreg_auc_scores.mean())

print('\nLinear SVC auc scores:', linearSVC_auc_scores)
print('Mean linear auc accuracy scores:', linearSVC_auc_scores.mean())

print('\nSVC auc scores:', SVC_auc_scores)
print('Mean SVC auc scores:', SVC_auc_scores.mean())

print('\nRandom forest auc scores:', RF_auc_scores)
print('Mean random forest auc scores:', RF_auc_scores.mean())
```

## Test Set

```python
# Wrap the linearSVC model in the CalibratedClassifiedCV estimator to enable predicted probabilites later on
# neccesary for AUC calculation
calib_linearSVC_model = CalibratedClassifierCV(linearSVC_model)
```

```python
# Fitting the classifiers to training data
logreg_model.fit(X_train, y_train)
calib_linearSVC_model.fit(X_train, y_train)
SVC_model.fit(X_train, y_train)
RF_model.fit(X_train, y_train)
```

```python
# Testing prediction accuracy
logreg_prediction = logreg_model.predict(X_test)
linearSVC_prediction = calib_linearSVC_model.predict(X_test)
SVC_prediction = SVC_model.predict(X_test)
RF_prediction = RF_model.predict(X_test)
```

```python
# Evaluating classifiers using accuracy score on test set
# Accuracy score is the ratio of correct predictions to total predictions
print(f'Logisitc Regression accuracy score is {accuracy_score(y_test, logreg_prediction)}')
print(f'Linear SVC accuracy score is {accuracy_score(linearSVC_prediction, y_test)}')
print(f'SVC accuracy score is {accuracy_score(SVC_prediction, y_test)}')
print(f'Random Forest accuracy score is {accuracy_score(RF_prediction, y_test)}')
```

```python
# Confusion matrices
print(confusion_matrix(logreg_prediction, y_test))
print(confusion_matrix(linearSVC_prediction, y_test))
print(confusion_matrix(SVC_prediction, y_test))
print(confusion_matrix(RF_prediction, y_test))
```

```python
prec_logreg = precision_score(y_test, logreg_prediction)
prec_linearSVC = precision_score(y_test, linearSVC_prediction)
prec_SVC = precision_score(y_test, SVC_prediction)
prec_RF = precision_score(y_test, RF_prediction)
print(f'Logisitc Regression precision score is {prec_logreg}')
print(f'Linear SVC precision score is {prec_linearSVC}')
print(f'SVCmacro precision is {prec_SVC}')
print(f'Random Forest precision score is {prec_RF}')
```

```python
# macro-F1 score
f1_logreg = f1_score(y_test, logreg_prediction, average='macro')
f1_linearSVC = f1_score(y_test, linearSVC_prediction, average='macro')
f1_SVC = f1_score(y_test, SVC_prediction, average='macro')
f1_RF = f1_score(y_test, RF_prediction, average='macro')
print(f'Logisitc Regression macro F1 score is {f1_logreg}')
print(f'Linear SVC macro F1 score is {f1_linearSVC}')
print(f'SVCmacro F1 score is {f1_SVC}')
print(f'Random Forest macro F1 score is {f1_RF}')
```

```python
# weighted-f1 score
f1_logreg = f1_score(y_test, logreg_prediction, average='weighted')
f1_linearSVC = f1_score(y_test, linearSVC_prediction, average='weighted')
f1_SVC = f1_score(y_test, SVC_prediction, average='weighted')
f1_RF = f1_score(y_test, RF_prediction, average='weighted')
print(f'Logisitc Regression weighted F1 score is {f1_logreg}')
print(f'Linear SVC weighted F1 score is {f1_linearSVC}')
print(f'SVC weighted F1 score is {f1_SVC}')
print(f'Random Forest weighted F1 score is {f1_RF}')
```

```python
# Graphing the weighted f1 score
# Classifier names
classifiers = ['Logistic Regression', 'Linear SVC', 'SVC', 'Random Forest']
# wieghted f1 scores for each classifier
f1_scores = [f1_logreg, f1_linearSVC, f1_SVC, f1_RF]
# Set up plot
fig, ax = plt.subplots()
# Create the bar plot
ax.bar(classifiers, f1_scores)
#Set y axis limits
ax.set_ylim([0, 1])
# Add labels to axes
ax.set_xlabel('Classifier')
ax.set_ylabel('Weighted F1-score')
# Annotate each bar with F1 value
for i, v in enumerate(f1_scores):
    plt.text(i, v+0.01, str(round(v, 4)), ha='center')
# Show plot
plt.show()
# Exporting graph
fig.savefig('F1_plot.png')
```

```python
# Getting predicted probabilities
y_pred_proba_logreg = logreg_model.predict_proba(X_test)[:, 1]
y_pred_proba_linearSVC = calib_linearSVC_model.predict_proba(X_test)[:, 1]
y_pred_proba_SVC = SVC_model.predict_proba(X_test)[:, 1]
y_pred_proba_RF = RF_model.predict_proba(X_test)[:, 1]
```

```python
# Calculating the false positive rate, the true positive rate and the threshold
fpr_logreg, tpr_logreg, thresholds_logreg = roc_curve(y_test, y_pred_proba_logreg)
fpr_linearSVC, tpr_linearSVC, thresholds_linearSVC = roc_curve(y_test, y_pred_proba_linearSVC)
fpr_SVC, tpr_SVC, thresholds_SVC = roc_curve(y_test, y_pred_proba_SVC)
fpr_RF, tpr_RF, thresholds_RF = roc_curve(y_test, y_pred_proba_RF)
# Calculate AUC-ROC using roc-auc function
roc_auc_logreg = auc(fpr_logreg, tpr_logreg)
roc_auc_linearSVC = auc(fpr_linearSVC, tpr_linearSVC)
roc_auc_SVC = auc(fpr_SVC, tpr_SVC)
roc_auc_RF = auc(fpr_RF, tpr_RF)

print(f'Logisitc Regression AUC is {roc_auc_logreg}')
print(f'Linear SVC AUC is {roc_auc_linearSVC}')
print(f'SVC AUC is {roc_auc_SVC}')
print(f'Random Forest AUC is {roc_auc_RF}')
```

```python
# Graphing the AUC for each classifier
# Classifier names
classifiers = ['Logistic Regression', 'Linear SVC', 'SVC', 'Random Forest']
# AUC for each classifier
AUC = [roc_auc_logreg, roc_auc_linearSVC, roc_auc_SVC, roc_auc_RF]
# Set up plot
fig, ax = plt.subplots()
# Create the bar plot
ax.bar(classifiers, AUC)
#Set y axis limits
ax.set_ylim([0, 1])
# Add labels to axes
ax.set_xlabel('Classifier')
ax.set_ylabel('AUC')
# Annotate each bar with AUC value
for i, v in enumerate(AUC):
    plt.text(i, v+0.01, str(round(v, 4)), ha='center')
# Show plot
plt.show()
# Exporting graph
fig.savefig('AUC_plot.png')
```

```python
# Plotting ROC curves for classifiers on same plot
plt.plot(fpr_logreg, tpr_logreg, label='Logistic Regression')
plt.plot(fpr_linearSVC, tpr_linearSVC, label='Linear SVC')
plt.plot(fpr_SVC, tpr_SVC, label='SVC')
plt.plot(fpr_RF, tpr_RF, label='Random Forest')
# Adding labels and legend
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
# Exporting graph
plt.savefig('ROC_plot.png')
# Showing graph
plt.show()
```

```python
# Plot the ROC curve for logistic regression model
plt.plot(fpr_logreg, tpr_logreg, color='darkorange', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc_logreg)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic - Logistic Regression')
plt.legend(loc="lower right")
plt.show()
```

```python
# Plot the ROC curve for linear SVC model
plt.plot(fpr_linearSVC, tpr_linearSVC, color='darkorange', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc_linearSVC)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic - Linear SVC')
plt.legend(loc="lower right")
plt.show()
```

```python
# Precision-Recall (AUC) values
from sklearn.metrics import average_precision_score, precision_recall_curve

precision_logreg, recall_logreg, _ = precision_recall_curve(y_test, y_pred_proba_logreg)
precision_linearSVC, recall_linearSVC, _ = precision_recall_curve(y_test, y_pred_proba_linearSVC)
precision_SVC, recall_SVC, _ = precision_recall_curve(y_test, y_pred_proba_SVC)
precision_RF, recall_RF, _ = precision_recall_curve(y_test, y_pred_proba_RF)

auprc_logreg = average_precision_score(y_test, y_pred_proba_logreg)
auprc_linearSVC = average_precision_score(y_test, y_pred_proba_linearSVC)
auprc_SVC = average_precision_score(y_test, y_pred_proba_SVC)
auprc_RF = average_precision_score(y_test, y_pred_proba_RF)

print('AUPRC_logreg:', auprc_logreg)
print('AUPRC_linearSVC:', auprc_linearSVC)
print('AUPRC_SVC:', auprc_SVC)
print('AUPRC_RF:', auprc_RF)
```

```python
# Plotting PR curves
plt.plot(precision_logreg, recall_logreg, label='Logistic Regression')
plt.plot(precision_linearSVC, recall_linearSVC, label='Linear SVC')
plt.plot(precision_SVC, recall_SVC, label='SVC')
plt.plot(precision_RF, recall_RF, label='Random Forest')
# Adding Legend
plt.legend()
# Setting axes titles
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()
```

## Model Complexity

```python
# The BIC of models will be compared. The AIC puts less emphasis than BIC on simple models so it will not be used.
import statsmodels.api as sm
from sklearn.metrics import log_loss
import math
```

```python
# Creating a list of the models
models = [logreg_model, calib_linearSVC_model, SVC_model, RF_model]

# Creating a list to store the AIC and BIC values
aic_values = []
bic_values = []

# Loop through the list of models
for model in models:
    # Fit the model
    model.fit(X_train, y_train)
    
    # Compute AIC and BIC of model
    aic = sm.regression.linear_model.RegressionResults.aic(model)
    bic = sm.regression.linear_model.RegressionResults.bic(model)
    
    # Append the values to the lists
    aic_values.append(aic)
    bic_values.append(bic)
    
# Display the AIC and BIC values for each model
for i in range(len(models)):
    print("Model", i+1, "AIC:", aic_values[i], "BIC:", bic_values[i])
```

```python
# BIC for logistic regression model
# bic = 2*LL+k*log(n)
log_likely_logreg = -log_loss(y_test, y_pred_proba_logreg)
# n is number of observations in the training data set
n = 293
# k for logreg is number of features + 1
k = 1074
bic_logreg = 2*(log_likely_logreg)+k*math.log(n)
print(bic_logreg)
```

```python
# BIC for linear SVC model
# bic = 2*LL+k*log(n)
log_likely_linearSVC = -log_loss(y_test, y_pred_proba_linearSVC)
# n is number of observations in the training data set
n = 293
# k for logreg is number of features + 1
k = 1074
bic_linearSVC = 2*(log_likely_linearSVC)+k*math.log(n)
print(bic_linearSVC)
```

```python
# BIC for SVC model
# bic = 2*LL+k*log(n)
log_likely_SVC = -log_loss(y_test, y_pred_proba_SVC)
# n is number of observations in the training data set
n = 293
# k for logreg is number of features + 1 (because SVC is using linear kernel)
k = 1074
bic_SVC = 2*(log_likely_SVC)+k*math.log(n)
print(bic_SVC)
```

```python
# Calculating number of parameters in random forest classifer
# Get number of leaves in all the trees
n_leaves = sum(tree.get_n_leaves() for tree in RF_model.estimators_)
# Get the number of input features
n_features = X_train.shape[1]
# Calculate number of parameters
n_params = n_leaves * n_features
print(n_params)
```

```python
# BIC for RF model
# bic = 2*LL+k*log(n)
log_likely_RF = -log_loss(y_test, y_pred_proba_RF)
# n is number of observations in the training data set
n = 293
# k for logreg is number of features + 1
k = n_params
bic_RF = 2*(log_likely_logreg)+k*math.log(n)
print(bic_RF)
```

```python
# Graphing the BIC for each classifier
# Classifier names
classifiers = ['Logistic Regression', 'Linear SVC', 'SVC', 'Random Forest']
# BIC for each classifier
BIC = [bic_logreg, bic_linearSVC, bic_SVC, bic_RF]
# Set up plot
fig, ax = plt.subplots()
# Create the bar plot
ax.bar(classifiers, BIC)
# Set y axis limits
# ax.set_ylim([0, 1])
# Add labels to axes
ax.set_xlabel('Classifier')
ax.set_ylabel('BIC')
ax.set_yscale('log')
# Annotate each bar with BIC value
for i, v in enumerate(BIC):
    plt.text(i, v+0.01, str(round(v, 3)), ha='center')
# Exporting graph
fig.savefig('BIC_plot.png')
# Show plot
plt.show()
```
