## This file converts MSA results from FASTA file into a matrix format and converts the MSA matrix into a binary matrix.

### Converting FASTA file into MSA Matrix

```python
from Bio import AlignIO
import pandas as pd
import numpy as np
```

```python
# Load MSA file in FASTA format
alignment = AlignIO.read('BVBRC_msa_refPA14.fasta', 'fasta')
```

```python
# Check number of sequences and length of alignment
num_sequences = len(alignment)
print(num_sequences)
alignment_length = alignment.get_alignment_length()
print(alignment_length)
```

```python
# Create dataframe to hold MSA matrix
MSA_df = pd.DataFrame()
```

```python
# Fill the dataframe with the sequences from the alignment
for i, record in enumerate(alignment):
    MSA_df[i] = list(record.seq)
```

```python
print(MSA_df)
```

```python
#Flip columns and rows
MSA_df = MSA_df.transpose()
MSA_df
```

### Creating Binary Matrix from MSA Matrix
#### This is done by comparing the nucleotide in each position to the reference strain. If same value = 0, if different value = 1)

```python
print(MSA_df.at[1,5])
```

```python
MSA_df.at[1,5].upper()
```

```python
if MSA_df.at[1,2].upper() == MSA_df.at[0, 2]:
    print("yes")
else:
    print("no")
```

```python
if MSA_df.at[1,5] == MSA_df.at[0, 5]:
    print("yes")
else:
    print("no")
```

```python
# # Testing binary conversion on row 1
# for i in range(MSA_df.shape[1]):
#     if MSA_df.at[1, i].upper() == MSA_df.at[0, i]:
#         MSA_df.at[1, i] = 0
#     else:
#         MSA_df.at[1, i] = 1
```

```python
# Binary conversion of entire matrix
for j in range(1, MSA_df.shape[0]):
    for i in range(MSA_df.shape[1]):
        if MSA_df.at[j, i].upper() == MSA_df.at[0, i]:
            MSA_df.at[j, i] = 0
        else:
            MSA_df.at[j, i] = 1
```

```python
MSA_df
```

```python
# Adding a column to dataframe for delta G
MSA_df['Delta G'] = MSA_df.sum(axis=1)
MSA_df
```

```python
MSA_df.at[0, 'Delta G'] = 0
MSA_df
```
